## Task 1 - Dealing with Null values and Duplicates

```python
# Checking for Duplicates
df.duplicated().sum()
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Finding the number of missing values in dataframe
df.isnull().sum().sum()
```

```python
# Count total number of rows with missing values
df.isna().any(axis=1).sum()
```

- There are no duplicate values in the dataframe
- There are 10 missing values in 10 distinct rows in the dataframe

```python
# Deleting Null Values
df = df.dropna(how='any')
df.info()
```

- Approach chosen to address the presence of null values is to remove the rows containing them. 
- Since 10 rows are less than 1% of the total number of rows in the data, removing them from the data will not have a strong effect on the results of the analysis to be performed

## Statistical Inferences for Each Column

### Fixed Acidity


```python
# General Statistical metrics for fixed acidity column
f_a = df['fixed acidity']
f_a.describe()
```

```python
# Histogram of fixed acidity 
sns.histplot(df['fixed acidity'])
```

```python
# Calculating the Mean fixed acidity
fa_mean = df['fixed acidity'].mean()
fa_mean
```

```python
#Checking for samples higher or lower than the average value
avg_compare = df['fixed acidity'] > fa_mean
avg_compare.value_counts()
```

```python
# Pie Chart showing amount of wine samples  measured against the average
plt.pie(avg_compare.value_counts(), 
labels=['Lower than Average', 'Higher Than Average'],
autopct='%1.1f%%')
plt.title(f'Pie Chart of Fixed Acidity content of Beverage stock(Average = {fa_mean})')
plt.legend(title="Fixed Acidity",
         loc="center left",
         bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```
